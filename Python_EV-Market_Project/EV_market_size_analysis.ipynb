```python
## TASK 2

# Already have the mean energy consumption column, so computed standard deviation directly. 
std_energy = df["mean - Energy consumption [kWh/100 km]"].std() 

# Compute Z-score using the existing mean column
df["Z_Score"] = df["mean - Energy consumption [kWh/100 km]"] / std_energy
 
 # Defining threshold for outliers
threshold = 2.5

# Filtering outliers
outliers = df[(df["Z_Score"] > threshold) | (df["Z_Score"] < threshold)]

# Rounding the Z_score to 2 decimal places 
outliers_display = outliers[['Car full name', "Z_Score", "mean - Energy consumption [kWh/100 km]"]].copy()
outliers_display["Z_Score"] = outliers_display["Z_Score"].round(2)
```

```python
# Display the EVs that are outliers
print("Outlier EVs:")
outliers_display
```

## **Insights**

#### 1. If the Z_score is higher than 2.5 then there is an EV which has the outlier among the EVs.

#### 2. Audi manufacturer has most cars in the outlier based on mean-Energy consumption.

```python
class EVRecommender:
    def __init__(self, ev_dataframe):
        """
        Initialize with a DataFrame containing EV data.
        :param ev_dataframe: Pandas DataFrame with EV details.
        """
        self.ev_database = ev_dataframe

    def recommend_evs(self, budget, desired_range, battery_capacity):
        """
        Recommend the top three EVs based on user criteria.
        :param budget: User's budget (PLN).
        :param desired_range: Minimum range required (km).
        :param battery_capacity: Minimum battery capacity required (kWh).
        :return: DataFrame with top three EV recommendations.
        """
        # Filter EVs based on criteria
        filtered_evs = self.ev_database[
            (self.ev_database['Minimal price (gross) [PLN]'] <= budget) &
            (self.ev_database['Range (WLTP) [km]'] >= desired_range) &
            (self.ev_database['Battery capacity [kWh]'] >= battery_capacity)
        ]

        # Sort by price (ascending) and return top 3 recommendations
        return filtered_evs.sort_values(by="Minimal price (gross) [PLN]").head(3)


if __name__ == "__main__":
    # Create the recommender system
    recommender = EVRecommender(df)

    # User inputs
    budget = int(input("Enter your budget (PLN): "))
    desired_range = int(input("Enter desired range (km): "))
    battery_capacity = int(input("Enter battery capacity (kWh): "))

    # Get recommendations
    recommendations = recommender.recommend_evs(budget, desired_range, battery_capacity)

    # Display results
    if recommendations.empty:
        print("No EVs match your criteria.")
    else:
        print("\nTop 3 EV Recommendations:")
        for idx, row in recommendations.iterrows():
            print(
                f"{row['Car full name']} - "
                f"Price: {row['Minimal price (gross) [PLN]']} PLN, "
                f"Range: {row['Range (WLTP) [km]']} km, "
                f"Battery: {row['Battery capacity [kWh]']} kWh"
            )
```

```python
import scipy.stats as stats

# Filter data for Tesla and Audi
tesla_engine_power = df[df["Make"] == "Tesla"]["Engine power [KM]"]
audi_engine_power = df[df["Make"] == "Audi"]["Engine power [KM]"]

# Perform Welch's t-test (unequal variances)
t_stat, p_value = stats.ttest_ind(tesla_engine_power, audi_engine_power, equal_var=False)

# Print results
print(f"\nT-test results: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis; There is a significant difference in engine power between Tesla and Audi.")
else:
    print("Fail to reject the null hypothesis; No significant difference in engine power between Tesla and Audi.")
```

## **Insights**

#### 1.No meaningful difference in engine power.

#### 2.Both brands should focus on other differentiators like range, batter capacity, acceleration, charging speed, or price to attract customers.
